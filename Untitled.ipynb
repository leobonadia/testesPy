```python
placas = pd.read_excel('placas.xlsx')

placas_sis = []
placas_sp = []

for placa in placas['placas']:
    if placa[-5] == '-':
        placa = f'{placa[:3]}{placa[-4:]}'
        placas_sis.append(placa)
    else:
        placas_sis.append(placa)
placas['placas_sis'] = placas_sis

for placa in placas['placas']:
    if placa[-3].isnumeric():
        if placa[-5] == '-':
            placas_sp.append(placa)
        else:
            if placa[-3].isnumeric():
                placa = f'{placa[:3]}-{placa[-4:]}'
                placas_sp.append(placa)   
    else:
        placa = f'{placa[:3]}{placa[-4:]}'
        placas_sp.append(placa)

placas['placas_sp'] = placas_sp
display(placas)

placas.to_excel('placasn.xlsx')
```
